```python
# Import necessary libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Function: Reduce noise using multiple methods
# This function applies a sequence of noise reduction methods to an image.
def reduce_noise(image, methods, kernel_size=5):
    """
    Reduce noise in a black-and-white image or CT scan using multiple methods.

    Parameters:
        image (numpy.ndarray): Input grayscale image.
        methods (list of str): List of noise reduction methods to apply ('median', 'gaussian', 'nlm').
        kernel_size (int): Kernel size for filtering (default is 5).

    Returns:
        denoised_image (numpy.ndarray): Denoised image after applying all methods.
    """
    denoised_image = image.copy()  # Create a copy of the original image to avoid overwriting

    # Apply each method in the list sequentially
    for method in methods:
        if method == "median":
            # Apply median filtering
            denoised_image = cv2.medianBlur(denoised_image, 23)
        elif method == "gaussian":
            # Apply Gaussian blurring
            denoised_image = cv2.GaussianBlur(denoised_image, (11, 11), 0)
        elif method == "nlm":
            # Apply Non-Local Means (NLM) denoising
            denoised_image = cv2.fastNlMeansDenoising(denoised_image, None, 10, 7, 21)
        else:
            # Raise an error if an unsupported method is specified
            raise ValueError(f"Unsupported method: {method}. Choose 'median', 'gaussian', or 'nlm'.")

    return denoised_image  # Return the final denoised image

# Title: Test the Noise Reduction Function
# Define the image path and methods to use
image_path = "/content/OPenCV_Basics_MEXE-4101_ATIENZA_MANALO/Medical Images/image_a.png"  # Replace with the actual image path
methods = ["median"]  # Add or modify methods as needed

try:
    # Load the original image in grayscale
    original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if original_image is None:
        raise FileNotFoundError(f"Image at {image_path} not found.")

    # Apply noise reduction
    denoised_image = reduce_noise(original_image, methods)

    # Title: Display Original and Denoised Images
    # Plot the original and denoised images side by side
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1), plt.imshow(original_image, cmap='Blues'), plt.title("Original")
    plt.subplot(1, 2, 2), plt.imshow(denoised_image, cmap='Blues'), plt.title(f"Denoised ({', '.join(methods)})")
    plt.show()
except FileNotFoundError as e:
    # Handle the case where the image file is not found
    print(e)
```
